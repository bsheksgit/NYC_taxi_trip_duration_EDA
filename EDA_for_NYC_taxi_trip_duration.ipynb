```python
# Calculating distance using latitudes and longitudes.
```

```python
from math import sin, cos, sqrt, atan2
```

```python
def calc_distance(lon1,lat1,lon2,lat2):
    R = 6373.0
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = (sin(dlat/2))**2 + cos(lat1) * cos(lat2) * (sin(dlon/2))**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance
```

```python
data['distance'] = data.apply(lambda x: calc_distance(x['pickup_longitude'],x['pickup_latitude'],x['dropoff_longitude'],x['dropoff_latitude']),axis=1)
data
```

```python
date_cols = ['day_of_month','month_number','day_of_year','day_of_week']
numerical_1 = ['passenger_count','distance']
time_cols = ['pickup_hour', 'dropoff_hour']
```

```python
def Univariate_numerical(data, var_group):
    size = len(var_group)
    x=1
    y=size
    plt.figure(figsize = (7*size,5), dpi = 100)
    for j,i in enumerate(var_group):
        mini = data[i].min()
        maxi = data[i].max()
        ran = data[i].max()-data[i].min()
        mean = data[i].mean()
        median = data[i].median()
        st_dev = data[i].std()
        skew = data[i].skew()
        kurt = data[i].kurtosis()

        # calculating points of standard deviation
        points = mean-st_dev, mean+st_dev

        #Plotting the variable with every information
        plt.subplot(x,y,j+1)
        sns.kdeplot(data[i], fill=True)
        sns.lineplot(x=points, y=[0,0], color = 'black', label = "std_dev")
        sns.scatterplot(x=[mini,maxi], y=[0,0], color = 'orange', label = "min/max")
        sns.scatterplot(x=[mean], y=0,color = 'red', label = "mean")
        sns.scatterplot(x=[median], y=0, color = 'blue', label = "median")
        plt.xlabel('{}'.format(i), fontsize = 20)
        plt.ylabel('density')
        plt.title('std_dev = {}; kurtosis = {};\nskew = {}; range = {}\nmean = {}; median = {}'.format((round(points[0],2),round(points[1],2)),
                                                                                                       round(kurt,2),
                                                                                                       round(skew,2),
                                                                                                       (round(mini,2),round(maxi,2),round(ran,2)),
                                                                                                       round(mean,2),
                                                                                                       round(median,2)))
```

```python
Univariate_numerical(data,numerical_1)
```

Conclusions:
1. We can see that the most number of trips had a single or two passengers
2. The distance coloumn seems to have a very large range. Hence the mean value is grater than the median.

Removing outliers and plotting some coloumns

```python
def Univariate_numerical_outliers(col):
   
    plt.figure(figsize = (7,5), dpi = 100)
    
    mini = data[col].min()
    maxi = data[col].max()
    ran = data[col].max()-data[col].min()
    mean = data[col].mean()
    median = data[col].median()
    st_dev = data[col].std()
    skew = data[col].skew()
    kurt = data[col].kurtosis()

    # calculating points of standard deviation
    points = mean-st_dev, mean+st_dev
    
    #Removing Outliers
    data_without_outliers = data[col][data[col] <= (median+3*st_dev)]
    
    #Plotting the variable with every information
    
    sns.kdeplot(data_without_outliers, fill=True)
#     sns.lineplot(points, color = 'black', label = "std_dev")
#     sns.scatterplot([mini,maxi], color = 'orange', label = "min/max")
#     sns.scatterplot([mean], color = 'red', label = "mean")
#     sns.scatterplot([median], color = 'blue', label = "median")
    plt.xlabel('{}'.format(col), fontsize = 20)
    plt.ylabel('density')
    plt.title('std_dev = {}; kurtosis = {};\nskew = {}; range = {}\nmean = {}; median = {}'.format((round(points[0],2),round(points[1],2)),
                                                                                                   round(kurt,2),
                                                                                                   round(skew,2),
                                                                                                   (round(mini,2),round(maxi,2),round(ran,2)),
                                                                                                   round(mean,2),
                                                                                                   round(median,2)))
```

```python
def UVA_category(data, var_group):

  # setting figure_size
  size = len(var_group)
  plt.figure(figsize = (7*size,5), dpi = 100)

  # for every variable
  for j,i in enumerate(var_group):
    norm_count = data[i].value_counts(normalize = True)
    n_uni = data[i].nunique()

  #Plotting the variable with every information
    plt.subplot(1,size,j+1)
    sns.barplot(x=norm_count, y=norm_count.index , order = norm_count.index)
    plt.xlabel('fraction/percent', fontsize = 20)
    plt.ylabel('{}'.format(i), fontsize = 20)
    plt.title('n_uniques = {} \n value counts \n {};'.format(n_uni,norm_count*100))
```

```python
data['vendor_id'] = data['vendor_id'].astype('category')
```

```python
categorical_vars = ['vendor_id','store_and_fwd_flag','month_name']
```

```python
UVA_category(data,categorical_vars)
```

1. 53% of the orders have vendor ID 2. But the difference between vendor ID 1 and 2 is small.
2. Store and fwd flag in almost all orders is 'N', which means there was no network issue in almost all rides.
3. March has the highest number of rides but there isn't much difference between all of the months.

```python
#removing outliers
def outlier_remover(variable,stddevs):
    median = data[variable].median()
    return data[data[variable]<median + (stddevs*data[variable].std())].loc[:,:]
```

```python
tripdur = outlier_remover('trip_duration',1.5)['trip_duration'].max()
rangearray = np.arange(0,tripdur,3000)
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
sns.scatterplot(x= data['passenger_count'], y= outlier_remover('trip_duration',1.5)['trip_duration'])
plt.yticks(rangearray)

# Taking the log graph for distance to get a better idea about the distribution.
plt.subplot(2,2,2)
sns.scatterplot(x=np.log(outlier_remover('distance',1.5)['distance']), y=np.log(outlier_remover('trip_duration',1.5)['trip_duration']))
plt.yticks(np.arange(0,np.log(tripdur),np.log(3000)))

plt.subplot(2,2,3)
sns.scatterplot(x=outlier_remover('pickup_hour',1.5)['pickup_hour'], y=outlier_remover('trip_duration',1.5)['trip_duration'])
plt.yticks(rangearray)

plt.subplot(2,2,4)
sns.scatterplot(x=outlier_remover('day_of_week',1.5)['day_of_week'], y=outlier_remover('trip_duration',1.5)['trip_duration'])
plt.yticks(rangearray)
```

```python
def multivariate_no_outliers_plot(data, categorical_list1, categorical_list2, numerical_list):
   
    
    sizer = len(categorical_list1)
    sizec = len(numerical_list)
    sizel2 = len(categorical_list2)
                
    plt.figure(figsize = (sizer*sizel2*10,sizec*10), dpi = 100)
    l=0
    for i in categorical_list2:
        for j in categorical_list1:
            for k in numerical_list:
                l+=1
                stdd = data[k].std()
                mean = data[k].mean()
                
                data_without_outliers = data[data[k] <= mean + (1.5*stdd)]
                
                #Plotting the variable with every information
                plt.subplot(sizer*sizel2,sizec,l)
                sns.boxplot(data_without_outliers, x=j, y= k, hue= i, orient='v')
                
```
